```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(font_scale=2)
```

```python
df = pd.read_csv("../data/raw/heart.csv")
print(df.shape)
df.head()
```

```python
df.describe()
```

```python
sns.countplot(x="target", data=df);
```

```python
sns.countplot(x="sex", data=df);
```

```python
df["age_interval"] = df["age"] // 10
```

```python
pd.crosstab(df["age_interval"], df["target"]).plot(kind="bar", figsize=(20, 8));
```

```python
for col in df:
    plt.figure(figsize=(20, 8))
    sns.histplot(df[col])
    plt.show()
```
